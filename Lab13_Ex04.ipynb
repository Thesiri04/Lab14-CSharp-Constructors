2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
var circle = new Circle();
var rectangle = new Rectangle();
var triangle = new Triangle();

class Shape
{
    private int? NumOfSide;
    private Shape()
    {
        System.Console.WriteLine("This is some shape with unknown side");
    }
    public Shape(int NumOfSide)
    {
        System.Console.WriteLine($"This is some shape with {NumOfSide} sides" );
    }
}
class Circle :Shape
{
    public Circle():base()
    {
        System.Console.WriteLine("This is a circle");
    }
}
class Rectangle :Shape
{
    public Rectangle(): base(4)
    {
        System.Console.WriteLine("This is a rectangle");
    }
}
class Triangle :Shape
{
   public Triangle() : base(3)
    {
        System.Console.WriteLine("This is a triangle");
    }
}
```

1. คลาส Shape:
มี 2 constructor:
- Shape() (protected): แสดงข้อความว่า "This is some shape with unknown side"
- Shape(int NumOfSide) (public): แสดงข้อความว่า "This is some shape with {NumOfSide} sides"
2. คลาสลูก (เช่น Circle, Rectangle, Triangle):
คลาสเหล่านี้จะสืบทอดมาจาก Shape และเรียกใช้ constructor ของ Shape โดยส่งค่าต่างๆ เช่น:
- Circle: เรียก Shape() และแสดง "This is a circle"
- Rectangle: เรียก Shape(4) และแสดง "This is a rectangle"
- Triangle: เรียก Shape(3) และแสดง "This is a triangle"
