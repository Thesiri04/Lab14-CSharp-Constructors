## Constructors

1.สร้าง console application project

```cmd
dotnet new console --name Lab13_Ex02
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
var circle = new Circle();
var rectangle = new Rectangle();
var triangle = new Triangle();

class Shape
{
    public Shape()
    {
        System.Console.WriteLine("This is some shape");
    }
}
class Circle :Shape
{
    public Circle()
    {
        System.Console.WriteLine("This is a circle");
    }
}
class Rectangle :Shape
{
    public Rectangle()
    {
        System.Console.WriteLine("This is a rectangle");
    }
}
class Triangle :Shape
{
   public Triangle()
    {
        System.Console.WriteLine("This is a triangle");
    }
}
```

6.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 5

7.อธิบายสิ่งที่พบในการทดลอง

1. คลาส Shape เป็นคลาสฐาน (Base Class)
- คลาส Shape มี Constructor ที่จะพิมพ์ "This is some shape" ทุกครั้งที่มีการสร้างอ็อบเจ็กต์จาก Shape หรือคลาสที่สืบทอดมาจาก Shape.
2. คลาส Circle, Rectangle, และ Triangle สืบทอดจาก Shape
- ทุกๆ คลาสนี้มี Constructor ที่จะพิมพ์ข้อความเฉพาะของคลาสนั้นๆ เช่น
- Circle → "This is a circle"
- Rectangle → "This is a rectangle"
+ Triangle → "This is a triangle"
3. การทำงานของ Constructor ในคลาสที่สืบทอด (Inheritance)
- เมื่อสร้างอ็อบเจ็กต์จากคลาสที่สืบทอดมาจาก Shape เช่น Circle, Rectangle, หรือ Triangle จะมีการเรียก Constructor ของ Shape ก่อนที่จะเรียก Constructor ของคลาสที่สืบทอด (นั่นคือ Circle, Rectangle, หรือ Triangle).
