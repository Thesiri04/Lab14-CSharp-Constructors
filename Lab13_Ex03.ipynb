## Constructors

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
var circle = new Circle();
var rectangle = new Rectangle();
var triangle = new Triangle();

class Shape
{
    private int? NumOfSide;
    public Shape()
    {
        System.Console.WriteLine("This is some shape with unknown side");
    }
    public Shape(int NumOfSide)
    {
        System.Console.WriteLine($"This is some shape with {NumOfSide} sides" );
    }
}
class Circle :Shape
{
    public Circle():base()
    {
        System.Console.WriteLine("This is a circle");
    }
}
class Rectangle :Shape
{
    public Rectangle(): base(4)
    {
        System.Console.WriteLine("This is a rectangle");
    }
}
class Triangle :Shape
{
   public Triangle() : base(3)
    {
        System.Console.WriteLine("This is a triangle");
    }
}
```

7.อธิบายสิ่งที่พบในการทดลอง

1. คลาส Shape มี 2 Constructor:
- Constructor ที่ไม่มีพารามิเตอร์: เมื่อถูกเรียก, จะพิมพ์ "This is some shape with unknown side".
- Constructor ที่รับพารามิเตอร์เป็น NumOfSide: เมื่อถูกเรียก, จะพิมพ์ข้อความที่ระบุจำนวนด้าน เช่น "This is some shape with 4 sides".
2. คลาส Circle, Rectangle, และ Triangle:
- Circle: เมื่อสร้างอ็อบเจ็กต์จากคลาส Circle, จะเรียก Constructor ของ Shape ที่ไม่มีพารามิเตอร์ (base()), แล้วพิมพ์ "This is some shape with unknown side", และตามด้วย "This is a circle".
- Rectangle: เมื่อสร้างอ็อบเจ็กต์จากคลาส Rectangle, จะเรียก Constructor ของ Shape ที่รับพารามิเตอร์ NumOfSide เป็น 4 (base(4)), แล้วพิมพ์ "This is some shape with 4 sides", และตามด้วย "This is a rectangle".
- Triangle: เมื่อสร้างอ็อบเจ็กต์จากคลาส Triangle, จะเรียก Constructor ของ Shape ที่รับพารามิเตอร์ NumOfSide เป็น 3 (base(3)), แล้วพิมพ์ "This is some shape with 3 sides", และตามด้วย "This is a triangle".
